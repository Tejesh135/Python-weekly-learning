Data Types in Python

```python
'''
data types in python
1. int
2. float
3. str
4. bool
5. list
6. tuple
7. dict
8. set
9. complex
10. bytes
11. frozenset
12. range
13. NoneType
'''

# int
a = 1
print(type(a))

# float
b = 1.0
print(type(b))

# str
c = "hi"
print(type(c))

# bool
d = True
print(type(d))

# list
e = [1,2,3]
print(type(e))

# tuple
f = (1,2,3)
print(type(f))

# dict
g = {1:"hi",2:"hello"}
print(type(g))

# set
h = {1,2,3}
print(type(h))

# complex
i = 1+2j
print(type(i))

# bytes
j = b"hi"
print(type(j))

# frozenset
k = frozenset({1,2,3})
print(type(k))

# range
l = range(1,2,3)
print(type(l))

# NoneType
m = None
print(type(m))

# type conversion
a = 1
b = float(a)
print(b)
print(type(b))

# int to str
a = 1
b = str(a)
print(b)
print(type(b))

# int to bool
a = 1
b = bool(a)
print(b)
print(type(b))

# int to list
a = 1,2,3
b = list(a)
print(b)
print(type(b))

# int to tuple
a = 1,2,3
b = tuple(a)
print(b)
print(type(b))


# int to complex
a = 1
b = complex(a)
print(b)
print(type(b))
```

Conditional Statements

```python
# Conditional Statements
'''
if
elif
else
'''

# if
a = 1
if a > 0:
    print("a is positive")

# elif
a = 0
if a > 0:
    print("a is positive")
elif a == 0:
    print("a is zero")
else:
    print("a is negative")

# else
a = -1
if a > 0:
    print("a is positive")
elif a == 0:
    print("a is zero")
else:
    print("a is negative")

# nested if
a = 1
b = 2
if a > 0:
    if b > 0:
        print("a and b are positive")
    else:
        print("a is positive and b is negative")
else:
    if b > 0:
        print("a is negative and b is positive")
    else:
        print("a and b are negative")

# short hand if
a = 1
if a > 0: print("a is positive")

# short hand if else
a = 1
print("a is positive") if a > 0 else print("a is negative")

# nested short hand if else
a = 1
b = 2
print("a is positive") if a > 0 else print("a is negative") if b > 0 else print("b is negative")

# and
a = 1
b = 2
if a > 0 and b > 0:
    print("a and b are positive")

# or
a = 1
b = 2
if a > 0 or b > 0:
    print("a or b is positive")

# not
a = 1
if not a > 0:
    print("a is not positive")

# pass
a = 1
if a > 0:
    pass
```
