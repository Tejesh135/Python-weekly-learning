```python
#if-elif
your_age = int(input("Enter your AGE: "))

if your_age >= 18:
    print("you can vote")
elif your_age <= 18:
    print("you cannot vote")
# hear if we enter 18+ first it will check cond1 if it is true then it will print or it will check 2nd one
```

```python
#if-elif-else
#lets see grade eg
your_marks = int(input("Enter your marks: "))

if your_marks >= 90:
    print("Grade A")
elif your_marks >= 70:
    print("Grade B")
else:
    print("Grade C")
```

```python
your_num = int(input("Enter your number: "))

if your_num % 2 == 0:
    print("even")
else:
    print("odd")
```

```python
# while looop
i = 2
while i <= 20:
    print(i)
    i += 2
# += is increment
```

```python
# odd no eg
i = 1
while i <= 20:
    print(i)
    i += 2
```

```python
n = int(input("Enter a number: "))
i = 1
total = 0

while i <= n:
    total += i
    i += 1

print("Sum =", total)
```
